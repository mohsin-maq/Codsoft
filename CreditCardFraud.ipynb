## **Feature Engineering**

```python

# Assuming you have a DataFrame df with a column 'trans_date_trans_time'
data[['datetime1', 'datetime2']] = data['trans_date_trans_time'].str.split(' ', 1, expand=True)

# Now df will have two new columns, 'datetime1' and 'datetime2'
```

```python
data.drop(columns='trans_date_trans_time',inplace=True)
```

```python
data=data[['amt', 'state',
       'zip', 'lat', 'long', 'city_pop', 'job',
        'merch_lat', 'merch_long', 'is_fraud',
       'datetime2']]
```

```python
data['job'].unique()
```

## **Ordinal Encoding**

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal_encoder=OrdinalEncoder()
columns_to_encode=['state','job']
data[columns_to_encode] = ordinal_encoder.fit_transform(data[columns_to_encode])
```

```python
data['datetime2'] = data['datetime2'].str.extract(r'(\d{2}:\d{2}:\d{2})')[0]
```

```python
def time_to_seconds(time_str):
    h, m, s = map(int, time_str.split(':'))
    return h * 3600 + m * 60 + s

# Apply the function to the 'time_column' and create a new column 'total_seconds'
data['total_seconds'] = data['datetime2'].apply(time_to_seconds)
```

```python
data.drop(columns="datetime2",inplace=True)
```

Now data looks good to be trained

## **Model Evaluation (Training & Testing)**

```python
print(data[data['is_fraud']==1.])
```

```python
data['is_fraud'].unique()
```

```python
data = data[data['is_fraud'] != 2.]
```

```python
data[data['is_fraud']==1.]
```

```python
data
```

```python
X=data.drop(columns="is_fraud")
y=data["is_fraud"]
```

## **Oversampling**

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy={1.:100000})

# Apply SMOTE to oversample the minority class "1"
X_resampled, y_resampled = smote.fit_resample(X, y)

# Check the class distribution after oversampling
print("Class distribution after SMOTE:\n", y_resampled.value_counts())
```

## **Train-test-split**

```python
X=X_resampled
y=y_resampled
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## **Standard Scaling**

```python
from sklearn.preprocessing import StandardScaler
```

```python
standard=StandardScaler()
X_train=standard.fit_transform(X_train)
X_test=standard.transform(X_test)
```

## **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
logi=LogisticRegression()
```

```python
logi.fit(X_train,y_train)
```

```python
y_pred = logi.predict(X_test)

```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Calculate accuracy
accuracy_1 = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy_1)

# Generate a classification report
report = classification_report(y_test, y_pred)
print("Classification Report:\n", report)

# Create a confusion matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:\n", cm)
```

## **ROC CURVE**

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_test, y_pred)

# Calculate the area under the ROC curve (AUC)
roc_auc = auc(fpr, tpr)

# Plot the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
custom_threshold = 0.4

# Get predicted probabilities for the positive class
y_pred2 = logi.predict_proba(X_test)[:,1]

# Apply the custom threshold to make binary predictions
y_custom_threshold = (y_pred2 >= custom_threshold).astype(int)
```

```python
# Calculate accuracy
accuracy = accuracy_score(y_test, y_custom_threshold)
print("Accuracy:", accuracy)

# Generate a classification report
report = classification_report(y_test, y_custom_threshold)
print("Classification Report:\n", report)

# Create a confusion matrix
cm = confusion_matrix(y_test, y_custom_threshold)
print("Confusion Matrix:\n", cm)
```

## ***DecisionTrees and  RandomForest***

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
decision_tree_model = DecisionTreeClassifier(random_state=42)
decision_tree_model.fit(X_train, y_train)

# Predictions using Decision Tree
y_pred_decision_tree = decision_tree_model.predict(X_test)

# Random Forest Classifier
random_forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
random_forest_model.fit(X_train, y_train)

# Predictions using Random Forest
y_pred_random_forest = random_forest_model.predict(X_test)

# Evaluate Decision Tree model
print("Decision Tree Classifier:")
print("Accuracy:", accuracy_score(y_test, y_pred_decision_tree))
print("Classification Report:\n", classification_report(y_test, y_pred_decision_tree))

# Evaluate Random Forest model
print("\nRandom Forest Classifier:")
print("Accuracy:", accuracy_score(y_test, y_pred_random_forest))
print("Classification Report:\n", classification_report(y_test, y_pred_random_forest))
```

## **Result**

```python
accuracies = [accuracy , accuracy_score(y_test, y_pred_decision_tree) ,  accuracy_score(y_test, y_pred_random_forest)]
models = ['LogisticRegression', 'DecisionTrees', 'RandomForest']

# Create a bar chart to compare accuracies
plt.figure(figsize=(8, 6))
sns.barplot(x=models, y=accuracies, data=data, palette='viridis')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Comparison of Models Based on Accuracy')
plt.ylim([0, 1])  # Set the y-axis limits to match the accuracy scale (0 to 1)
plt.xticks(rotation=45)  # Rotate the x-axis labels for readability
plt.tight_layout()
plt.show()






```
